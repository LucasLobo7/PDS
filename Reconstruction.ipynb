```python
import os
import imp
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
import sys
import pickle
import matplotlib.ticker as plticker
from scipy.signal import find_peaks
from numpy.lib.scimath import sqrt as csqrt
plt.rcParams["figure.figsize"] = (15,7)


```

```python
tinicial = -2
tfinal = 2
Namostras = 40
Ncontinuo = 1000000
plotsincs = True
Nsincs = 10
plotpontos = True

t = np.linspace(tinicial,tfinal,Ncontinuo)
n = np.linspace(tinicial,tfinal,Namostras)

y = np.sin(2*np.pi*n) + np.cos(np.pi*n)

T = (tfinal-tinicial)/(Namostras-1)

analogico = np.zeros(Ncontinuo)
analogicotemp = np.zeros(Ncontinuo)

if plotpontos:
    #plt.stem(n,y, basefmt="")
    plt.plot(n,y,'o')



sincs = np.linspace(tinicial,tfinal,Nsincs)
indice = 0
for i in range(Namostras):
    analogicotemp = y[i]*np.sinc((t-n[i])/T)
    analogico = analogico + analogicotemp
    if plotsincs and (i == indice):
        plt.plot(t,analogicotemp,linestyle='--')
        indice += Namostras/Nsincs 


#plt.plot(n,y) #discreta
plt.plot(t,analogico) # analogica
#plt.plot(t,np.sin(2*np.pi*t) + np.cos(np.pi*t)) # "continua"
#plt.xlim(0.5,1)
```
